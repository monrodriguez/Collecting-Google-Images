```python
#Will keep scrolling down the webpage until it cannot scroll no more
last_height = driver.execute_script('return document.body.scrollHeight')
while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    time.sleep(2)
    new_height = driver.execute_script('return document.body.scrollHeight')
    try:
        driver.find_element_by_xpath('//*[@id="islmp"]/div/div/div/div/div[5]/input').click().screenshot(folder+str(searchterms)+'.png')
        time.sleep(2)
    except:
        pass
    if new_height == last_height:
        break
    last_height = new_height
```

```python
#number of images to collect
number=20
```

```python
URLs = []
Texts=[]
Pics=[]
numbers=[]
folders=[]
Dates=[]

count=0
for i in range(1,number):
    try:
        image=driver.find_element('xpath', '//*[@id="islrg"]/div[1]/div['+str(i)+']/a[1]/div[1]/img')
        image.screenshot(downloadfolder+str(searchterms)+str(i)+'.png')
        
        Pic=str(searchterms)+str(i)+'.png'
        
        info=driver.find_element('xpath', '//*[@id="islrg"]/div[1]/div['+str(i)+']/a[2]')
        URL=info.get_attribute("href")
        Text=info.get_attribute("title")       
        
        num=i
        fold=downloadfolder
        time.sleep(2)
        count+=1
        
        today = date.today()
        
        print(num)
        print (fold)
        print(Pic)
        print (URL)
        print (Text)
        print (today)

        URLs.append(URL)
        Texts.append(Text)
        Pics.append(Pic)
        numbers.append(num)
        folders.append(fold)
        Dates.append(today)

    except:
        pass
```

```python
zipped = list(zip(numbers, folders, Dates, Pics, Texts, URLs))
```

```python
df = pd.DataFrame(zipped, columns=['number', 'folder', 'Date', 'Pic', 'Text', 'URL'])
```
